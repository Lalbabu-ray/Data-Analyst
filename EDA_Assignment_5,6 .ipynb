### QUESTION 5B-->FOR ALL THE COUNTRIES INVOLVED,FOR ALL THE SALES IN 2013,WHICH FLAVORS DOMINATED THE MARKET IN THE YEAR 2013

```python
def topfiveflavors(df):

  d={}
  for i in df['flavor']:
    
    l=[]

    l=i.split(';')
    for j in l:
      if j not in d:

        d[j]=1

      else:

        d[j]=d[j]+1

  value_key_pairs = ((value, key) for (key,value) in d.items())
  sorted_value_key_pairs = sorted(value_key_pairs, reverse=True)
  

  for i in range(0,6):

    print(sorted_value_key_pairs[i])



```

```python
topfiveflavors(df_product_2013)
```

```python

  
def topfiveposition(df1,df2):  
  d={}
  l=[]
  for i in df1['positioning']:

      

      l.extend(i.split(','))

  for j in l:

      for k in range(0,len(df2)):
        
        if df2['Positioning Subcategory'].iloc[k]==j:

          if df2['Positioning Group'].iloc[k] not in d:

            d[df2['Positioning Group'].iloc[k]]=1

          else:

            d[df2['Positioning Group'].iloc[k]]=d[df2['Positioning Group'].iloc[k]]+1
      
  

  value_key_pairs = ((value, key) for (key,value) in d.items())
  sorted_value_key_pairs = sorted(value_key_pairs, reverse=True)

  
  res=[]
  for i in range(0,5):

     res.extend(sorted_value_key_pairs[i])

  
  return res





  







   






```

```python
topfiveposition(df_product_2013,position_df)
```
