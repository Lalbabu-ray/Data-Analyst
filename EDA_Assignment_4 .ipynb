```python
df_product['eventmonth']=pd.DatetimeIndex(df_product['eventdate']).month
df_product['eventmonth']
```

```python


for i in range(0,len(df_product)):
  
  if df_product['eventmonth'].iloc[i]<4:
    df_product['eventquarter'].iloc[i]=1
  
  elif 6>=df_product['eventmonth'].iloc[i]>=4:

    df_product['eventquarter'].iloc[i]=2

  elif 9>=df_product['eventmonth'].iloc[i]>=7:

    df_product['eventquarter'].iloc[i]=3

  else:

    df_product['eventquarter'].iloc[i]=4


  
  
```
